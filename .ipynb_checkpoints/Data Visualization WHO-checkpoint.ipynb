```python
# Test
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('who-data.csv')
data['MA'] = data['New_cases'].rolling(window=7,min_periods=1).mean()
data['MA_deaths'] = data['New_deaths'].rolling(window=7, min_periods=1).mean()
ger = data[data.Country == 'Germany']
fr = data[data.Country == 'France']
at = data[data.Country == 'Austria']
pl = data[data.Country == 'Poland']
cz = data[data.Country == 'Czechia']
ch = data[data.Country == 'Switzerland']

data.head(2)
```

```python
plt.figure(figsize=(15,10))
plt.grid(False)
plt.plot(ger.Date_reported,fr['MA_deaths'], 'r--')
plt.plot(pl.Date_reported,pl['MA_deaths'], 'c--')
plt.plot(cz.Date_reported,cz['MA_deaths'], 'm--')
plt.plot(ger.Date_reported,ger['MA_deaths'], 'b-')
plt.plot(ch.Date_reported,ch['MA_deaths'], 'k--')
plt.plot(at.Date_reported,at['MA_deaths'], 'g--')


plt.xticks([])
plt.ylabel('Todesfälle (7-Tage Mittel)', fontsize=20)
plt.title('Todesfälle pro Tag (7-Tage Mittel), Deutschland und Nachbarn, WHO-Daten via Python', fontsize=24)
plt.xlabel('Zeit', fontsize=20)
plt.legend(['Frankreich', 'Polen', 'Tschechien', 'Deutschland', 'Schweitz', 'Österreich'], fontsize=20)

plt.savefig('plot_deaths.jpg', dpi = 300)
```

```python
plt.figure(figsize=(15,10))
plt.grid(False)
plt.plot(ger.Date_reported,fr['MA'], 'r--')
plt.plot(pl.Date_reported,pl['MA'], 'c--')
plt.plot(ger.Date_reported,ger['MA'], 'b-')
plt.plot(cz.Date_reported,cz['MA'], 'm--')
plt.plot(ch.Date_reported,ch['MA'], 'k--')
plt.plot(at.Date_reported,at['MA'], 'g--')

plt.xticks([])
plt.ylabel('Neue Fälle (7-Tage Mittel)', fontsize=20)
plt.xlabel('Zeit', fontsize=20)
plt.title('Neue Fälle pro Tag (7-Tage Mittel), Deutschland und Nachbarn, WHO-Daten via Python', fontsize=24)

plt.legend(['Frankreich', 'Polen', 'Deutschland', 'Tschechien', 'Schweitz', 'Österreich'], fontsize=20)

plt.savefig('plot_cases.jpg', dpi = 300)
```
