```python
import pandas as pd #Daten
from matplotlib import pyplot as plt # plots
#import matplotlib as mpl
from matplotlib.ticker import FuncFormatter   # Numberformat plot ticks
import numpy as np

# performance
from datetime import datetime
from datetime import date # todays date
#import seaborn as sns

import os
now = datetime.now()
today = date.today().strftime("%d.%m.%Y")
```

```python
# Webabruf - CSV einlesen
df = pd.read_csv("https://covid19.who.int/WHO-COVID-19-global-data.csv")
```

```python
ger = df[df.Country == 'Germany']
fr = df[df.Country == 'France']
at = df[df.Country == 'Austria']
pl = df[df.Country == 'Poland']
cz = df[df.Country == 'Czechia']
ch = df[df.Country == 'Switzerland']
it = df[df.Country == 'Italy']
es = df[df.Country == 'Spain']
gb = df[df.Country == 'The United Kingdom']

isr = df[df.Country == 'Israel']
se = df[df.Country == 'Sweden']
usa = df[df.Country == 'United States of America']
spa = df[df.Country == 'Spain']
ita = df[df.Country == 'Italy']
rus = df[df.Country == 'Russian Federation']
```

```python
plt.style.use('seaborn')


plt.figure(figsize=(16,9))


y1 = ger["New_cases"].rolling(window=7,min_periods=1).mean()
x1 = ger["Cumulative_cases"]

y2 = gb["New_cases"].rolling(window=7,min_periods=1).mean()
x2 = gb["Cumulative_cases"]

y3 = usa["New_cases"].rolling(window=7,min_periods=1).mean()
x3 = usa["Cumulative_cases"]

y4 = isr["New_cases"].rolling(window=7,min_periods=1).mean()
x4 = isr["Cumulative_cases"]

# y = gb["New_cases"].rolling(window=7,min_periods=1).mean()
# x = gb["Cumulative_cases"]

plt.plot(x1,y1, label = "Deutschland", linewidth=5)
plt.plot(x2,y2, label = "GroÃŸbritannien")
plt.plot(x3,y3, label = "USA")
plt.plot(x4,y4, label = "Israel")

plt.yscale('log')
plt.xscale('log')

plt.ylabel('Neue FÃ¤lle', fontsize = 20)
plt.xlabel('kumulierte FÃ¤lle', fontsize=20)

plt.title('Neue FÃ¤lle pro Tag - absolute Zahlen (WHO-Daten)\n', fontsize=20)
plt.suptitle(today + ' PW', fontsize=15, y=0.91)


plt.legend(loc='upper center', 
           bbox_to_anchor=(0.5, -0.2),
           fancybox=True, 
           shadow=True, 
           ncol=3, 
           fontsize=25)

plt.savefig("D:\\OneDrive\\Auswertung_Covid_19\\CumulativeVsNewCases.png", dpi = 200, bbox_inches='tight')


plt.show()
```
